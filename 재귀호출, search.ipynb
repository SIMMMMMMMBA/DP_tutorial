# 1. 재귀함수란?

간단하게 말하면, 재귀함수는 함수가 자기 자신을 호출하는 함수를 말합니다.

재귀 호출에 대한 비유 출처: https://www.shutterstock.com

재귀호출은 계속 함수가 반복되는것이 아니라, 위 그림의 문이 점점 멀어져가듯이 함수가 호출이 되면서 점점 깊이 들어가는 구조입니다. 즉 반복문으로 함수를 연속 호출하는 것과는 차이점이 있습니다.

자세하게 설명하기보단 예제를 통하여 알아보도록 하겠습니다.

1~n까지의 합을 for문으로 구현한다면

```python
tot = 0 
n = 10
for i in range(1,n+1):
    tot +=i
print(tot)
```

이렇게 코딩을 할 수 있습니다. 하지만 1~n까지의 합을 재귀적으로 생각해 봅시다.

n + n-1 + .... + 1 = n까지의 합

n + (n-1)까지의 합 = n까지의 합

으로 나타낼 수 있습니다. f(n)은 1~n까지의 합이라고 할 때 이 식을 일반화 한다면,

f(n) = n + f(n-1) 으로 나타낼 수 있습니다. 

```python
def f(n):
    if(n==1):
        return 1
    return n+f(n-1)
print(f(10))
```

위는 이를 코드로 나타낸 것입니다. 다만, 보시는 것과 같이

if(n==1):
    return 1

처럼 중간에 STOP POINT를 정해주지 않으면, 무한번 반복하게 되어 문제가 생깁니다. 즉, 재귀함수에서는 STOP POINT를 설정하는 것과 문제를 재귀적으로 생각하는 것이 중요합니다.

#### 1) 유클리드 호제법
두 정수 a, b의 최대공약수를 G(a, b)라고 하자.

정수 a, b, q r (b ≠ 0)에 대하여 a = bq + r,이면 G(a, b) = G(b, r)가 성립한다.

〈증명〉
G(a, b) = g라고 하자. 최대공약수의 성질에 의해 a = a′g, b = b′g이고 G(a′, b′) = 1이다.
a = bq + r로부터 r = a - bq = g(a′ - b′q) 이고, g는 r의 약수이다.
G(b, r) = g임을 보이기 위해서는 G(b′, a′ - b′q) = 1임을 보이면 된다.

귀류법을 적용하여 결론을 부정해보자.
어떤 정수 d가 존재하여 G(b′, a′ - b′q) =d(≠ 1)라고 하면, b′ = dm, a′ - b′q = dn라고 할 수 있고, a′ = b′q + dn = dmq + dn = d(mq + n) 이므로 G(a′, b′) = 1라는 가정에 모순이다. 따라서 G(b′, a′ - b′q) = 1이므로 G(b, r) = g가 성립한다.

#### 2) 유클리드 호제법의 사용
유클리드 호제법은 큰 수들의 최대공약수를 구할 때 사용할 수 있다.

a = bq + r 에서 r = a - bq 이므로 G(a, b) = G(a, a - bq)이다. 이제 직접 적용을 해보자.
(예) G(180, 200) = G(180, 200 - 180) = G(180, 20) = G(180 - 20 × 8, 20) = G(20, 20) = 20

일반적으로 두 정수의 최대공약수를 구할 때에는 소인수분해1)하여 공통인수를 찾아낸다. 하지만 소인수분해할 때, 합성수가 큰 소수를 인수로 갖는다면 소인수분해가 쉽지 않을 것이다. 유클리드 호제법은 소인수분해하지 않고 최대공약수를 구할 수 있는 방법이다.
